```python
import os

root_dir = "./Speakers/train/"

def list_not_hidden(directory):
    return [f for f in os.listdir(directory) if not f.startswith(('.','_')) ]

for speaker in list_not_hidden(root_dir):
    speaker_dir = os.path.join(root_dir, speaker)
    print(f"processing -- {speaker_dir}")

    for f in list_not_hidden(speaker_dir):
        if f.endswith(".mp3"):
            mp3_path = os.path.join(speaker_dir, f)
            wav_path = os.path.join(speaker_dir, f"{os.path.splitext(f)[0]}.wav")
            if not os.path.isfile(wav_path):
                AudioSegment.from_mp3(mp3_path).export(wav_path, format="wav")
                print(f"converting {mp3_path} to {wav_path}")

print("Done!")
```

### Dataset

After all the files are converted and stored in the structured directories, you can use torch `DatasetFolder` to create your dataset.

`DatasetFolder` is an interesting function, it can easily parse your train directory and infer the number of classes, and while iterating over files it can also apply a specific function to each file. In our case, we want to loop over the files and convert them to numeric format. We can do this by sending each file to the following __loader()__ function. 

```python
def loader(path):
    audio, sample_rate = torchaudio.load(path)
    audio = torchaudio.functional.resample(audio, orig_freq=sample_rate, new_freq=16000)
    audio = audio.mean(dim=0, keepdim=True)
    mfcc_features = torchaudio.transforms.MFCC(sample_rate=sample_rate, n_mfcc=13, melkwargs={"n_mels": 40})(audio)
    mfcc_features = mfcc_features.mean(dim=2).squeeze()
    return mfcc_features
```

```python
loader(wav_path).shape, loader(wav_path)
```

<div class="alert-info">
The above loader function reads an audio file, resamples it to 16K and extracts MFCC features from it, then takes the average of the MFCC features and returns a vector of dimension 13. You can see this vector as a new representation of your audio file.
</div>

Create the __train_dataset__ by calling the `DatasetFolder` function with the train directory path as __root__, "loader" for the __loader__, and "wav" as __extensions__.

```python
from torchvision.datasets import DatasetFolder
train_dataset = DatasetFolder(root='Speakers/train', loader=loader, extensions=("wav",))
```

### Train/Validation Split

Split your dataset into a training dataset and a validation dataset.

```python
# Automatically determine split sizes
train_size = int(0.8 * len(train_dataset))  # 80% for training
val_size = len(train_dataset) - train_size  # 20% for validation

# Correct way to split
train_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])

# Check dataset sizes
print(f"Training samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")
```

### DataLoaders

Use the created __train_dataset__, __val_dataset__ to build the __train_loader__ and __val_loader__. Select an appropriate batch_size. 

```python
train_loader = DataLoader(dataset=train_dataset, batch_size=32)
val_loader = DataLoader(dataset=val_dataset, batch_size=32)
```

```python
len(train_loader), len(val_loader)
```

### Building a Model

Create a fully connected neural network. You can choose the number of layers, neurons, and activation functions.

```python
class SpeakerClassifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()

        # Fully connected layers
        self.fc1 = nn.Linear(13, 256)  # Change input size to 13 (since loader gives 13 features)
        self.fc2 = nn.Linear(256, 128)  # Hidden layer
        self.fc3 = nn.Linear(128, num_classes)  # Output layer

        # Activation function
        self.relu = nn.ReLU()

    def forward(self, x):
        if x.dim() == 2:  # Ensure input is correct shape (batch_size, 13)
            x = x.view(x.size(0), -1)  # Keep (batch_size, 13)

        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)  # No activation for output (CrossEntropyLoss will handle it)
        return x
```

### Model Training 

You know the rest! 

Bring in the accuracy function and the training loop. 

Train the model and plot the loss and accuracy curves.

```python
import torch

def accuracy_fn(y_true, y_pred):
    correct = (torch.eq(y_true, y_pred)).sum().item()
    acc = (correct / y_true.size(0)) * 100  # Convert to percentage
    return acc
```

```python
num_classes = 34  
model = SpeakerClassifier(num_classes=num_classes).to(device)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0005)


train_losses, train_accs = [], []
validation_losses, validation_accs = [], []

num_epochs = 50  

for epoch in range(num_epochs):
   
    model.train()
    train_loss, train_acc = 0, 0
    
    for x_batch, y_batch in train_loader: 
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        
        optimizer.zero_grad()
        y_logits = model(x_batch)
        loss = loss_fn(y_logits, y_batch)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        y_pred = torch.argmax(y_logits, dim=1)  
        train_acc += accuracy_fn(y_true=y_batch, y_pred=y_pred) 
        
  
    train_losses.append(train_loss / len(train_loader))
    train_accs.append(train_acc / len(train_loader))

   
    model.eval()
    validation_loss, validation_acc = 0, 0  

    with torch.no_grad():
        for x_val, y_val in val_loader:
            x_val, y_val = x_val.to(device), y_val.to(device)
            y_logits = model(x_val)

            loss = loss_fn(y_logits, y_val)  # 
            validation_loss += loss.item()   

            y_pred = torch.argmax(y_logits, dim=1)  # here i coonvert logits to predicted labels
            validation_acc += accuracy_fn(y_true=y_val, y_pred=y_pred) 

 
    validation_loss /= len(val_loader)
    validation_acc /= len(val_loader)


    validation_losses.append(validation_loss)
    validation_accs.append(validation_acc)

    # i print to see the results step by step 
    if epoch % 10 == 0:
        print(f"Epoch: {epoch} | "
              f"Train Loss: {train_losses[-1]:.5f}, Train Acc: {train_accs[-1]:.2f}% | "
              f"Val Loss: {validation_losses[-1]:.5f}, Val Acc: {validation_accs[-1]:.2f}%")
```

```python
import matplotlib.pyplot as plt

# i wrote this code to ensure lists have the correct length because i had some error before with my code before! 
assert len(train_losses) == len(validation_losses), "Error: Train and Validation losses are not the same length!"
assert len(train_accs) == len(validation_accs), "Error: Train and Validation accuracies are not the same length!"

# Plot Loss Curve
plt.figure(figsize=(10, 4))
plt.plot(range(len(train_losses)), train_losses, label="Train Loss", marker='o')
plt.plot(range(len(validation_losses)), validation_losses, label="Validation Loss", marker='s')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss Curve")
plt.grid(True)
plt.show()

# Plot Accuracy Curve
plt.figure(figsize=(10, 4))
plt.plot(range(len(train_accs)), train_accs, label="Train Accuracy", marker='o')
plt.plot(range(len(validation_accs)), validation_accs, label="Validation Accuracy", marker='s')
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.legend()
plt.title("Accuracy Curve")
plt.grid(True)
plt.show()
```
